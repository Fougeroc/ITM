## Rauzy graph

```python
from itm import *
from itertools import permutations
vertices = []
for top in permutations('abc'):
    for bot in permutations('abc'):
        if top[0] != bot[0] and top[-1] != bot[-1] and top[0] != bot[1]:
            vertices.append(LabelledPermutationOver((' '.join(top), ' '.join(bot))))

G = DiGraph(loops=True,multiedges=True)
for v in vertices:
    for move in v.moves():
        G.add_edge(v, v.rauzy_move(move), move)
```

```python
def parity(path):
    n = len(path)/2
    r = 0
    for i in range(n):
        move = path[2*i:2*(i+1)]
        if move == 't1' or move == 't2':
            r += 1
    return r%2
```
