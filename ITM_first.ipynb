```python
def rauzy_path(E):
    if E.is_rauzy(-1):
        print 'R'
        E.rauzy(-1)
    elif E.is_rauzy(0):
        print 'L'
        E.rauzy(0)
    else:
        print 'S'
        E.rotation(E._lengths[0])
        rauzy_path(E)
```
